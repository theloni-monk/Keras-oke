```python
corrected_transcript = []
lyric_idx = 0
target_locs = []
LOOKAHEAD = 6
for script_segment in transcript_estimate:
    seg_words = script_segment[2].strip()
    seg_words = ''.join([i for i in seg_words if i.isalpha() or i == ' ']).split(' ')
    seg_embedding = c2v_model.vectorize_words(seg_words) # sequence of vectors of dimension 50

    num_words = len(seg_words)
    lookahead = min(num_words, LOOKAHEAD)

    srange_min = max(0,lyric_idx - 5)
    srange_max = min(lyric_idx + 20, len(lyrics))

    #compute embedding distance between sequences in the lyric and the estimate sequence
    correlation = []
    for i in range(len(lyrics)-lookahead):
        embedding_distance = np.linalg.norm(lyrics_embedding[i:i+lookahead] - seg_embedding[:lookahead])
        correlation.append(embedding_distance)
    correlation = [max(correlation) - i for i in correlation]
    correlation = np.array(correlation)

    # look for when the segment phrase starts
    try:
        amax = np.argmax(correlation[srange_min:srange_max]) # 3 is a fudge factor for how off the words could be
        target_location = srange_min + np.argmax(correlation[srange_min:srange_max])
    except ValueError:
        target_location = lyric_idx
    target_locs.append(target_location)
    
    print("\nseg_words", seg_words, "lyric_idx", lyric_idx, "\ntarget_location", target_location, "lyrics at target_location", lyrics[target_location:target_location+num_words])

    lyric_idx+=num_words

for i in range(0,len(target_locs)-1):
    script_segment = transcript_estimate[i]
    prev_target_loc = target_locs[i]
    target_loc = target_locs[i+1]
    target_lyrics = lyrics[prev_target_loc:target_loc] #TODO: don't use script_segment length as a reference for how many words
    corrected_transcript.append([script_segment[0], script_segment[1], target_lyrics])
corrected_transcript.append([script_segment[0], script_segment[1], lyrics[target_locs[-1]:len(lyrics)]])
```

```python
def format_timestamp(seconds: float, always_include_hours: bool = False, decimal_marker: str = '.'):
    assert seconds >= 0, "non-negative timestamp expected"
    milliseconds = round(seconds * 1000.0)

    hours = milliseconds // 3_600_000
    milliseconds -= hours * 3_600_000

    minutes = milliseconds // 60_000
    milliseconds -= minutes * 60_000

    seconds = milliseconds // 1_000
    milliseconds -= seconds * 1_000

    hours_marker = f"{hours:02d}:" if always_include_hours or hours > 0 else ""
    return f"{hours_marker}{minutes:02d}:{seconds:02d}{decimal_marker}{milliseconds:03d}"

# write transcript to closed caption file format for web
def write_vtt(transcript: list, file):
    print("WEBVTT\n", file=file)
    for segment in transcript:
        print(
            f"{format_timestamp(segment[0])} --> {format_timestamp(segment[1])}\n"
            f"{ ' '.join(segment[2]).replace('-->', '->')}\n",
            file=file,
            flush=True,
        )

# write transcript to closed caption file format
def write_srt(transcript: list, file):

    for i, segment in enumerate(transcript, start=1):
        # write srt lines
        print(
            f"{i}\n"
            f"{format_timestamp(segment[0], always_include_hours=True, decimal_marker=',')} --> "
            f"{format_timestamp(segment[1], always_include_hours=True, decimal_marker=',')}\n"
            f"{' '.join(segment[2]).strip().replace('-->', '->')}\n",
            file=file,
            flush=True,
        )
```

```python
def editDistance(r, h):
    '''
    This function is to calculate the edit distance of reference sentence and the hypothesis sentence.

    Main algorithm used is dynamic programming.

    Attributes: 
        r -> the list of words produced by splitting reference sentence.
        h -> the list of words produced by splitting hypothesis sentence.
    '''
    d = np.zeros((len(r)+1)*(len(h)+1), dtype=np.uint8).reshape((len(r)+1, len(h)+1))
    for i in range(len(r)+1):
        d[i][0] = i
    for j in range(len(h)+1):
        d[0][j] = j
    for i in range(1, len(r)+1):
        for j in range(1, len(h)+1):
            if r[i-1] == h[j-1]:
                d[i][j] = d[i-1][j-1]
            else:
                substitute = d[i-1][j-1] + 1
                insert = d[i][j-1] + 1
                delete = d[i-1][j] + 1
                d[i][j] = min(substitute, insert, delete)
    return d

def wer(r, h):
    """
    This is a function that calculate the word error rate in ASR.
    You can use it like this: wer("what is it".split(), "what is".split()) 
    """
    # build the matrix
    d = editDistance(r, h)

    result = float(d[len(r)][len(h)]) / len(r) * 100
    result = str("%.2f" % result) + "%"

    print("WER: " + result)
```
